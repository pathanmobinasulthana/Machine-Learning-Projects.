```python
outliers=[]
for i in newdf.columns:
    for x in newdf[i]:

        m=(newdf[i]).mean()
        sd=(newdf[i]).std()
        z = 0 if sd == 0 else ((x-m) / sd)
        if abs(z) > 3 :
                outliers.append(i)
print(set(outliers))
```
