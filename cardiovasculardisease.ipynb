## Columns with high correlation with target variable.


```python
for x in df.columns:
    cv=df[x].corr(df["target"])
    print(x,cv)
```

```python
highcorrelatedcolumns=["cp","thalach","exang","oldpeak"]
```

```python
df["target"].unique()
```

```python
dfx=df.drop("target",axis=1)
```

```python
df.x=df[highcorrelatedcolumns]
df.y=df["target"]
```

## Standradising the dataset.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit(df.x)
```

```python
encodeddf=sc.transform(df.x)
encodeddf
```

```python
scaleddf=pd.DataFrame(encodeddf)
```

```python
scaleddf.columns=df.x.columns
```

##  Splitting the data into train and test parts.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaleddf,df.y,test_size=0.2,random_state=123)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Building a LogisticRegression Model

```python
from sklearn import linear_model
```

```python

reg=linear_model.LogisticRegression()
```

## fitting the data into model and Predicting for xtest.

```python
reg.fit(x_train,y_train)
```

```python
ytestpredict=reg.predict(x_test)
```

```python
ytestpredict
```

## Building a RandomForest model to compare different models.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
yrf=rf.predict(x_test)
```

```python
yrf
```

## accuracy score of logistic regression

```python

from sklearn.metrics import accuracy_score
```

```python
lgas=accuracy_score(y_test,ytestpredict)
```

```python
lgas
```

## accuracy score of RandomForestClassifier

```python

accuracy_score(y_test,yrf)
```
